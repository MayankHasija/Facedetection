```python
import cv2
from cv2 import imread
from cv2 import imshow
from cv2 import waitKey
from cv2 import destroyAllWindows
from cv2 import CascadeClassifier
from cv2 import rectangle
import matplotlib
print(cv2.__version__)
```

```python
# load the pre-trained model|
classifier = cv2.CascadeClassifier(r'C:\Users\mayan\Desktop\1TouchSolutions\Face_recog\haarcascade_frontalface_default.xml')
```

```python
# load the photograph
pixels = cv2.imread(r'C:\Users\mayan\Desktop\1TouchSolutions\Face_recog\Face_test6.jpg',1)
pixels2 = cv2.cvtColor(pixels, cv2.COLOR_BGR2RGB)
```

```python
import matplotlib
from matplotlib import pyplot as plt
plt.imshow(pixels2)
plt.xticks([]), plt.yticks([])  # Hides the graph ticks and x / y axis
plt.show()
```

```python
# perform face detection
bboxes = classifier.detectMultiScale(pixels2)
# print bounding box for each detected face
for box in bboxes:
	# extract
	x, y, width, height = box
	x2, y2 = x + width, y + height
	# draw a rectangle over the pixels
	rectangle(pixels2, (x, y), (x2, y2), (0,0,255), 3)
```
